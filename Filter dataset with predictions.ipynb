```python
import pandas as pd
```

```python
pred_df = pd.read_csv('part_of_speech_estimated.csv', index_col=0)
pred_df
```

```python
label_df = pd.read_csv('datasets/wyandot words/labeled.tsv', sep='\t', index_col=0)
label_df
```

```python
og_df = pd.read_csv('tn_words_tabular.tsv', sep='\t', index_col=0)
og_df
```

```python
preds = pd.concat([label_df, pred_df.rename({'pred': 'morphology'}, axis=1)])
preds
```

```python
og_df.join(preds.loc[:, 'morphology'])
```

```python
verbs_df = og_df.join(preds.loc[:, 'morphology'])
verbs_df = verbs_df[~verbs_df.morphology.isna()]
mask = verbs_df.morphology.apply(lambda x: 'VERB' in x)
verbs_df = verbs_df[mask]
verbs_df
```

```python
verbs_df.to_csv('wyandot_verbs.tsv', sep='\t')
```
